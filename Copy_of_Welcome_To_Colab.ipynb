````
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
````

````
!pip install speechbrain pydub

import torchaudio
from speechbrain.pretrained import SpectralMaskEnhancement
from pydub import AudioSegment
import matplotlib.pyplot as plt
import IPython.display as ipd
````

````
# Load SpeechBrain's speech enhancement model
enhancer = SpectralMaskEnhancement.from_hparams(
    source="speechbrain/metricgan-plus-voicebank",
    savedir="pretrained_models/metricgan-plus-voicebank"
)
````

````
# Example: Load a noisy speech sample from your dataset
noisy_file = "/content/dataset/noisy_testset_wav/noisy_testset_wav/p232_001.wav"  # change this path

# Load the audio file to get data and sampling rate
try:
    waveform, sample_rate = torchaudio.load(noisy_file)

    # Listen to noisy speech
    print(f"Loading audio from: {noisy_file}")
    print(f"Sample rate: {sample_rate}")
    print(f"Waveform shape: {waveform.shape}")

    ipd.display(ipd.Audio(waveform.squeeze().numpy(), rate=sample_rate))

except FileNotFoundError:
    print(f"Error: The file {noisy_file} was not found. Please check the path.")
except Exception as e:
    print(f"An error occurred while loading or playing the audio: {e}")
````

````
import torch
# Load audio
noisy_waveform, fs = torchaudio.load(noisy_file)

# Convert the waveform to float32
noisy_waveform = noisy_waveform.to(torch.float32)

# Enhance (denoise)
# The enhance_batch method requires a 'lengths' argument.
# For a single audio file, the length is 1.
enhanced = enhancer.enhance_batch(noisy_waveform, lengths=torch.ones(1))


# Save enhanced audio
torchaudio.save("enhanced_speech.wav", enhanced.cpu(), fs)

print(" Enhanced speech saved as enhanced_speech.wav")
ipd.Audio("enhanced_speech.wav")
````

````
# Plot waveforms
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(noisy_waveform[0].numpy())
plt.title("Noisy Speech")

plt.subplot(1, 2, 2)
plt.plot(enhanced[0].detach().cpu().numpy())
plt.title("Enhanced Speech")

plt.show()
````
